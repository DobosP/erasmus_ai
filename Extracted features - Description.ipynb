## Stuff added

Based on: https://www.kaggle.com/wrosinski/baselinemodeling

### Sentiment metadata

* sentiment_sentiment_document_magnitude <-- got it already
* sentiment_sentiment_document_score    <-- got it already
* sentiment_sentiment_score_MEAN        <- it's linear to the overall doc score above (=useless)
* sentiment_sentiment_magnitude_MEAN    <- it's linear to the overall doc magnitude above (=useless)
* sentiment_sentiment_score_SUM         <- to add (done)
* sentiment_sentiment_magnitude_SUM     <- to add (done)
* sentiment_entities                    <- labels (words) - to add (done) - gets Vectorized later

### Image metadata

Each picture has:
      
* 'annots_score': file_top_score, <----- MEAN score of the 30% most meaningful labels (done)
* 'color_score': file_color_score, <---- MEAN score of all recognized colors on the picture (done)
* 'color_pixelfrac': file_color_pixelfrac, <---- MEAN pixelfrac of all recognized colors on the picture (done)
* 'crop_conf': file_crop_conf, <---- MEAN confidence of cropHints (objects I guess) / we have just 1st (done)
* 'crop_importance': file_crop_importance, <---- MEAN importance of cropHints / we have just 1st (done)
* 'annots_top_desc': self.sentence_sep.join(file_top_desc) <-- 30% of top labels (words) - gets Vectorized later
        

Per each PetID the features above are aggregated in the form of MEAN & SUM.
